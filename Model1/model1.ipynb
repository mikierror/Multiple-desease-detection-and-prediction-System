```python
X= data_set.drop(columns=['corona_result', 'test_date', 'test_indication'],axis=1)
print(X)
```

```python
X.head()
```

```python
Y= data_set ['corona_result']
Y.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X['age_60_and_above'] = X['age_60_and_above'].map({'No': 0, 'Yes': 1})
```

```python
X.head()
```

```python
X['gender'] = X['gender'].map({'male': 0, 'female': 1})
```

```python
X.head()
```

```python
scalar = StandardScaler()
scalar.fit(X)
standardized_data= scalar.transform(X)
print(standardized_data)
```

```python
X = standardized_data
Y = data_set['corona_result']
print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_Test,Y_Train,Y_Test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=1)
print(X.shape,X_train.shape,X_Test.shape)
```

```python
from sklearn import svm
```

```python
classifier = svm.SVC(kernel='linear')
classifier.fit(X_train,Y_Train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
# Make predictions on the training data
X_train_predictions = classifier.predict(X_train)

# Calculate accuracy on the training data
training_accuracy = accuracy_score(X_train_predictions, Y_Train)
print("Accuracy on training data:", training_accuracy)
```

```python
X_Test_prediction=classifier.predict(X_Test)
testing_data_prediction= accuracy_score(X_Test_prediction,Y_Test)
print("Accuracy on testing data : ",testing_data_prediction)
```
